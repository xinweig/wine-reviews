```python
import pandas as pd
import seaborn as sns
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv')
```

```python
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] <160], kind='hex', 
              gridsize=20, color='r', ).set_axis_labels("Wine Price", "Wine Points")
```
