```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv')
```

```python
reviews.points.describe()
```

```python
reviews.price.describe()
```

```python
fig, axarr = plt.subplots(2, 1, figsize=(12, 10))

reviews['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0], fontsize=12,)
axarr[0].set_title("Wine Points", fontsize=18)

reviews['price'].value_counts().plot.hist(bins = 80, range=(0,2000),
    ax=axarr[1], fontsize=12,)
axarr[1].set_title("Wine Price", fontsize=18)

plt.subplots_adjust(hspace=0.3)
```
